```
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
from pylab import *
import numpy as np
```

```
idades = [13,12,16,22,84,43,23,22,10,9,77,55,49]
```

```
ids = [ x for x in range(len(idades))]
```

```
# bar e a fun√ß√£o para fazer graficos de barras
plt.bar(ids, idades)
plt.show()
```

```
bins = [0,5,10,15,20,25,30,35,40,45,50,55,60]
```

```
# hist e a fun√ß√£o para fazer histogramas
plt.hist(idades, bins, histtype= 'bar', rwidth = 0.8)
plt.show()
```

```
plt.hist(idades, bins, histtype= 'stepfilled', rwidth= 0.8)
plt.show()
```

```
fatias = [6,4,4,10]
atividades = [ ' comer', ' dormir', 'estudar', 'descansar']
colunas = ['c', 'm', 'r', 'k']
```

```
plt.pie(fatias, labels = atividades, colors = colunas, startangle = 90, shadow = True, explode = (0,0.1,0,0))
plt.show()
```

```
x = linspace (3,13,23)
y = x**2

fig = plt.figure()

#definindo os eixos
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x,y,'r')

axes.set_xlabel1('x')
axes.set_ylabel1('y')
axes.set_title('Gr√°fico de linha');
```

```
fig, axes = plt.subplots(nrows= 1, ncols=2)

for ax in axes:
  ax.plot(x,y, 'r')
  ax.set_xlabel('x')
  ax.set_ylabel('y')
  ax.set_title('T√≠tulo')

fig.tight_layout()
```

```
# arange = array, random = valores randomicos que s√£o valores aleatorios
plt.scatter(np.arange(50), np.random.randn(50))
plt.show()
```

```
fig = plt.figure()

ax1 = fig.add_subplot(1,2,1)
ax1.plot(np.random.randn(50), color = 'yellow')

ax2 = fig.add_subplot(1,2,2)
ax2.scatter(np.arange(50), np.random.randn(50))
plt.show()
```

```
#grid coloca os quadrinhos
fig, axes = plt.subplots(1,2, figsize=(10,4))

axes[0].plot(x,x**2, x, exp(x))
axes[0].set_title("Escala Padr√£o")
axes[0].grid(True)

axes[1].plot(x,x**2,x,exp(x))
axes[1].set_yscale("log")
axes[1].set_title("Escala Logaritmica (y)");
```
